## <ins>Introduction</ins>

This project is about Neural style transfer for creating artistic images from regular images.

Style transfer is an optimization technique used to take two images — a content image and a style reference image (such as an artwork by a famous painter)—and blend them together so the output image looks like the content image, but “painted” in the style of the style reference image.

#### Creating a function to load images from path

```python
import cv2
import os

# Function to load images from folder
def load_images(folder):
  
  images = [] #Empty list to store the images

  for filename in os.listdir(folder):

    img = cv2.imread(os.path.join(folder,filename))

    # Keeping the color of images to RGB
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    if img is not None:

      images.append(img)

  # Preprocessing the images
  images = np.array(images)# Converting image list to numpy array

  images = images/255 #Scaling the images.
  """
  The pixel value of each image ranges between 0 and 255
  Dividing each image by 255 will scale the values between 0 and 1. 
  This is also known as normalization.

  """

  return images
```

### <ins>VGG Model</ins>

For this project, we are going to use the VGG19 model.

#### What is VGG19?

VGGNet is invented by VGG (Visual Geometry Group) from University of Oxford.

VGG19 is a variant of VGG model which in short consists of 19 layers (16 convolution layers, 3 Fully connected layer, 5 MaxPool layers and 1 SoftMax layer). There are other variants of VGG like VGG11, VGG16 and others. VGG19 has 19.6 billion FLOPs.

The intermediate layers of this model will be used for further processing.

#### Why intermediate layers?

According to the original [paper](https://arxiv.org/pdf/1409.1556.pdf), the intermediate outputs within the pretrained image classification network defines style and content representations.



```python
content_layers = ['block5_conv2'] 

style_layers = ['block1_conv1',
                'block2_conv1',
                'block3_conv1', 
                'block4_conv1', 
                'block5_conv1']

num_content_layers = len(content_layers)
num_style_layers = len(style_layers)
```

We need the content of our image to be prevalant over the style.

The lower layers of the model learns about the low level details such as shaders, textures, edge details etc.

Wheras the higher layers learns the actual content of an image such as the face, position of eyes etc.

### Building the Model

The following function builds a VGG19 model using Functional API that returns a list of intermediate layer outputs.

```python
def vgg19_layers(layers):

  #Loading the vgg19 model
  vgg = VGG19(include_top = False, weights = 'imagenet')

  #Since this model is already trained, we don't need to train it again.
  vgg.trainable = False

  # Gathering the layers
  output = [vgg.get_layer(name).output for name in layers]

  # Creating model
  model = tf.keras.Model([vgg.input], output)

  return model
```

Now, we need to calculate a gram matrix to calculate the style.

#### **What is gram matrix?**

[**Gram matrix**](https://medium.com/@cdsjatin/tell-me-again-how-much-gram-is-in-the-matrix-71652095b2c7#:~:text=Gram%20matrix%20is%20simply%20the,computed%20using%20the%20gram%20matrix.) is simply the matrix of the inner product of each vector and its corresponding vectors in same.

In deep learning , while style transferring the loss function is computed using the gram matrix.

The formula can be represented as:

<img src = "https://miro.medium.com/max/304/1*cb1VrP8Gnx7lf7uZHj-wSw.png" >

<img src = "https://miro.medium.com/max/2026/1*eAcdLzoo_8nnv1g3NTTPzg.png" >

This can be implemented using the **tf.linalg.einsum** function.

```python
def gram_matrix(input_tensor):

  result = tf.linalg.einsum('bijc,bijd->bcd', input_tensor, input_tensor) # tf.linalg.einsum allows defining Tensors by defining their element-wise computation.

  input_shape = tf.shape(input_tensor)

  num_locations = tf.cast(input_shape[1]*input_shape[2], tf.float32)

  return result/(num_locations)
```

#### Defining a model that will return style and content tensors

This model when called on an image will return the gram matrix of the style_layers and content of the content layers

```python
def Model(style_layers, content_layers, image):

  vgg = vgg19_layers(style_layers + content_layers)
  vgg.trainable = False

  num_style_layers = len(style_layers) #Total number of style layers

  #The model expects the image pixels to be in the range 0 and 1
  image = image*255.0 #Rescaling pixels back

  preprocessed_image = tf.keras.applications.vgg19.preprocess_input(image)# Preprocessing the inputs

  preprocessed_image = tf.image.resize(preprocessed_image, (256,256))#vgg19 model takes input images in (256,256) shape

  output = vgg(preprocessed_image)

  #Seperating representations of styles and contents
  style_outputs, content_outputs = (output[ : num_style_layers], output[ num_style_layers : ])

  style_outputs = [gram_matrix(style_out) for style_out in style_outputs]# Calculate the gram matrix for each layer in the style output.

  # Storing the styles and contents in the form of dictionary
  dict_content = {content : name for content, name in zip(content_layers, content_outputs)}
  dict_styles = {styles : name for styles, name in zip(style_layers, style_outputs)}

  return {'content_details': dict_content, 'style_details': dict_styles}
```

#### Calling the function on the content image and style image and extracting respective dictionaries.

```python
content_targets = Model(style_layers, content_layers, single_image)['content_details']
style_targets = Model(style_layers, content_layers, style_image)['style_details']
```

We'll store the single image as a tensorflow variable for later use.

```python
content_image = tf.Variable(single_image)
```

**The pixel values of the images need to be scaled again between 0 and 1**

Let's define a function to do that

```python
def scale(image):
  return tf.clip_by_value(image, clip_value_min=0.0, clip_value_max=1.0)
```

#### Next, we need to define our losses for styles and images.

```python
def losses(outputs):

  #Setting weights for individual losses for further optimization
  style_weight=1e-2
  content_weight=1e4

  style_outputs = outputs['style_details']
  content_outputs = outputs['content_details']

  #Setting Style loss
  style_loss = tf.add_n([tf.reduce_mean((style_outputs[name]-style_targets[name])**2) 
                          for name in style_outputs.keys()])
  style_loss *= style_weight / num_style_layers

  #Setting content loss
  content_loss = tf.add_n([tf.reduce_mean((content_outputs[name]-content_targets[name])**2) 
                            for name in content_outputs.keys()])
  content_loss *= content_weight / num_content_layers

  loss = style_loss + content_loss
  return loss
```


**Let's break the approach of the above code a bit.**

First, individual **weights** for styles and contents are initialized to compute total variation loss at later stage and further optimization.
Then the styles and content dictionaries for their respective layers are gathered.

Next, the **losses** are calculated seperately for styles and contents.
The difference between the styles of two images is the **style loss** as the difference between the activations of the style filters for each image. Same thing for the difference between content of two images (**content loss**), it’s the difference between the activations of the filters in the content layers of each image.

Finally, the **total loss** is computed.




Now , we have to define a tensorflow function that will apply style and update the image.

We will use tf.GradientTape()

Also, one downside to this basic implementation is that it produces a lot of high frequency artifacts. Decrease these using an explicit regularization term on the high frequency components of the image. In style transfer, this is often called the total variation loss.

```python
optimizer =  tf.keras.optimizers.Adam(lr = 0.01, beta_1=0.99, epsilon=1e-1)

total_variation_weight = 30

@tf.function()
def training(image):

  with tf.GradientTape() as tape:

    outputs = Model(style_layers, content_layers, content_image)
    loss = losses(outputs)
    loss += float(total_variation_weight*tf.image.total_variation(image))

  grad = tape.gradient(loss, image)

  optimizer.apply_gradients([(grad, image)])

  image.assign(scale(image))
```

The following function will be used to display the converted image.

```python
def tensor_to_image(tensor):

  tensor = tensor*255 #Scaling images back to original pixels

  tensor = np.array(tensor, dtype=np.uint8)# Changing tensor datatype to uint8

  if np.ndim(tensor)>3:

    assert tensor.shape[0] == 1
    tensor = tensor[0]

  return PIL.Image.fromarray(tensor)
```

Visit [here](https://pillow.readthedocs.io/en/stable/reference/Image.html#PIL.Image.fromarray) for more details about PIL.Image.fromarray

#### Finally training the model

```python
import time
start = time.time()

epochs = 100

for n in range(epochs):

  training(content_image)
  print(".", end='')

display.clear_output(wait=True)
display.display(tensor_to_image(content_image))

end = time.time()
print("Total time: {:.1f}".format(end-start))
```
